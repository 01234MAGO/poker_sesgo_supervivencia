#### ID DE MANO

```python
def extraccion_id_mano(texto):
    patron = r"Mano n.º (\d+)"
    coincidencias = re.findall(patron, texto)
    if coincidencias:
        return coincidencias[0]
    return None
```

#### ID DE TORNEO

```python
def extraccion_id_torneo(texto):
    pattern = r'Torneo n.º (\d+)'
    match = re.search(pattern, texto)
    if match:
        return match.group(1)
    else:
        return None
```

#### ENTRADA

```python
def extraccion_entrada(texto):
    match = re.search(r"Freeroll", texto)
    if match:
        return "freeroll"
    else:
        match = re.search(r"\d+\.\d+ ?\$[+/] ?\d+\.\d+ \$", texto)
        if match:
            return match.group()
        else:
            return None
```

#### NIVEL TORNEO

```python
def extraccion_nivel_torneo(texto):
    match = re.search(r"Nivel\s*(\w{1,9})", texto)
    if match:
        return match.group(1)
    else:
        return None
```

####  CIEGAS

```python
def extraccion_ciegas(texto):
    regex = r"\((\d+)/(\d+)\)"
    match = re.search(regex, texto)
    if match:
        return f"{match.group(1)}/{match.group(2)}"
    else:
        return None
```

### FUNCION BLOQUE 1

```python
def bloque1(texto):
    # Definimos la sentencia límite
    limite = "Asiento"

    # Inicializamos las variables para almacenar los datos extraídos
    id_mano = extraccion_id_mano(texto[:texto.find(limite)])
    tipo_juego = extraccion_tipo_juego(texto[:texto.find(limite)])
    id_torneo = extraccion_id_torneo(texto[:texto.find(limite)])
    entrada = extraccion_entrada(texto[:texto.find(limite)])
    nivel_torneo = extraccion_nivel_torneo(texto[:texto.find(limite)])
    ciegas = extraccion_ciegas(texto[:texto.find(limite)])
    fecha_hora = extraccion_fecha_hora(texto[:texto.find(limite)])
    id_mesa = extraccion_id_mesa(texto[:texto.find(limite)])
    max_jugadores = extraccion_max_jugadores(texto[:texto.find(limite)])
    asiento_boton = extraccion_asiento_boton(texto[:texto.find(limite)])

    # Devolvemos los datos extraídos separados por coma
    return f"{id_mano},{tipo_juego},{id_torneo},{entrada},{nivel_torneo},{ciegas},{fecha_hora},{id_mesa},{max_jugadores},{asiento_boton}"
```

```python
def extraccion_sube_cuanto(texto):
    regex = r"JaviBazte: sube\s*[\d\.]+\D*([\d\.]+)"
    matches = re.findall(regex, texto)
    if not matches:
        return None
    if len(matches) == 1:
        return float(matches[0])
    total = 0
    for match in matches:
        segundo = float(match)
        total += segundo
    return round(total, 2)
```

```python
def extraccion_bote_total(texto):
    sentencia = 'Bote total'
    valor_num = None

    posicion_sentencia = texto.find(sentencia)
    if posicion_sentencia != -1:
        resto_str = texto[posicion_sentencia+len(sentencia):]
        valor_str = ''
        for word in resto_str.split():
            if word.replace('.', '', 1).isdigit():
                valor_str = word
                break
        if valor_str:
            valor_num = float(valor_str) if '.' in valor_str else int(valor_str)

    return valor_num
```

#### CARTAS COMUNITARIAS

```python
def extraccion_cartas_comunitarias(texto):
    sentencia = 'Comunitarias'
    inicio_corchete = texto.find(sentencia)
    if inicio_corchete == -1:
        return None
    else:
        inicio_corchete = texto.find('[', inicio_corchete)
        fin_corchete = texto.find(']', inicio_corchete)
        if inicio_corchete == -1 or fin_corchete == -1:
            return None
        else:
            return texto[inicio_corchete+1:fin_corchete]
```

```python
def extraccion_bote_ganado(texto):
    patron = r'JaviBazte.*?(?=Asiento|$)'
    texto_jb = re.search(patron, texto, re.DOTALL)
    if texto_jb:
        texto_jb = texto_jb.group()
        match_gano = re.search(r'\b(ganó|recaudó)\b\D*(\d[\d,.]*)', texto_jb, re.I)
        if match_gano:
            numero = match_gano.group(2)
            numero = re.sub(r'[^\d.]', '', numero)
            return float(numero)
    return None
```

```python
def EDD(texto):
    # Ejecución del bloque1 desde el comienzo del texto
    bloque1_result = bloque1(texto)

    # Volver al principio del texto y buscar la sentencia 'botón'
    boton_start = texto.find('botón')

    # Ejecución del bloque2 desde la ubicación de 'botón'
    bloque2_result = bloque2(texto[boton_start:])

    # Volver al principio del texto y buscar la sentencia '*** CARTAS DE MANO ***'
    cartas_start = texto.find('*** CARTAS DE MANO ***')

    # Ejecución del bloque3 desde la ubicación de '*** CARTAS DE MANO ***'
    bloque3_result = bloque3(texto[cartas_start:])

    # Volver a la ubicación de '*** CARTAS DE MANO ***' y ejecutar el bloque4
    bloque4_result = bloque4(texto[cartas_start:])

    # Volver a las ubicaciones de '*** CARTAS DE MANO ***' y '*** RESUMEN ***' y ejecutar el bloque5
    resumen_start = texto.find('*** RESUMEN ***')
    bloque5_result = bloque5(texto[resumen_start:])

    # Combinación de los resultados de los bloques
    return f"{bloque1_result},{bloque2_result},{bloque3_result},{bloque4_result}{bloque5_result}"


```

```python
def contar_subtextos(ruta_archivo):
    patron = r"\*{11} n\. (\d+) \*{11}"
    contador_subtextos = 0

    with open(ruta_archivo, 'r') as archivo:
        contenido = archivo.read()

        # Buscar todas las coincidencias de la sentencia utilizando expresiones regulares
        coincidencias = re.findall(patron, contenido)

        # Contar el número de coincidencias encontradas
        contador_subtextos = len(coincidencias)

    return contador_subtextos


num_subtextos = contar_subtextos(ruta_archivo)

```

```python
def subtexto_numero(ruta_archivo):
    patron = r"\*{11} n\. (\d+) \*{11}"
    num_elementos = []

    with open(ruta_archivo, 'r') as archivo:
        contenido = archivo.read()

        # Dividir el contenido en subtextos utilizando expresiones regulares
        coincidencias = re.findall(patron, contenido)

        # Obtener el índice de inicio y fin de cada subtexto
        indices_inicio = [m.start() for m in re.finditer(patron, contenido)]
        indices_fin = [m.start() for m in re.finditer(patron, contenido)][1:] + [len(contenido)]

        # Obtener los subtextos a partir de los índices y ejecutar la función EDD
        for inicio, fin in zip(indices_inicio, indices_fin):
            subtexto = contenido[inicio:fin].strip()
            resultado_EDD = EDD(subtexto)  # Ejecutar la función EDD dentro del subtexto
            num_elementos.append(len(resultado_EDD.split(',')))

    return num_elementos

num_elementos = subtexto_numero(ruta_archivo)

# Imprimir el número de elementos separados por coma en cada resultado de la función EDD
for i, num_elementos_subtexto in enumerate(num_elementos):
    print(f"Subtexto {i+1}: Número columnas = {num_elementos_subtexto}")
```
